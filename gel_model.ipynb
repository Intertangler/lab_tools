## define or detect background

```python
#######################################################################################
#### select Bezier curve points for background subtraction - do by trial and error ####
#######################################################################################
bg_range_beginning = 5
bg_range_end = 550
bg_range_mid = 145
#######################################################################################


def recta(x1, y1, x2, y2):
  ### credit: https://stackoverflow.com/users/13389746/chema ###
    a = (y1 - y2) / (x1 - x2)
    b = y1 - a * x1
    return (a, b)

def curva_b(xa, ya, xb, yb, xc, yc, points):
  ### credit: https://stackoverflow.com/users/13389746/chema ###
    (x1, y1, x2, y2) = (xa, ya, xb, yb)
    (a1, b1) = recta(xa, ya, xb, yb)
    (a2, b2) = recta(xb, yb, xc, yc)
    puntos = []

    for i in range(0, points):
        if x1 == x2:
            continue
        else:
            (a, b) = recta(x1, y1, x2, y2)
        x = i*(x2 - x1)/points + x1
        y = a*x + b
        puntos.append((x,y))
        x1 += (xb - xa)/points
        y1 = a1*x1 + b1
        x2 += (xc - xb)/points
        y2 = a2*x2 + b2
    return puntos

lista1 = curva_b(bg_range_beginning, intensities[bg_range_beginning], bg_range_mid, intensities[bg_range_mid], bg_range_end, intensities[bg_range_end],bg_range_end-bg_range_beginning)

plt.plot(positions,intensities)
plt.scatter([bg_range_beginning,bg_range_mid,bg_range_end],[intensities[bg_range_beginning],intensities[bg_range_mid],intensities[bg_range_end]], c= "r")
plt.plot(*zip(*lista1), c='r', linestyle="--")
plt.xlabel('position [px]')
plt.ylabel('intesity [a.u.]')
plt.show()
new_intensities = intensities[bg_range_beginning:bg_range_end] - list(zip(*lista1))[1]
bg_subtracted_intensities = np.concatenate((np.concatenate((np.zeros((bg_range_beginning)),new_intensities)),np.zeros((len(intensities)-bg_range_end))))
plt.plot(positions, bg_subtracted_intensities)
plt.xlabel('position [px]')
plt.ylabel('intesity [a.u.]')
plt.show()
```

## fit function that predicts mobility as a function of length

```python
import scipy.optimize as opt
### gel parameters: 
run_time = 70*60 # min*60
gel_length = 7.5 #cm/100
E_field = 100/gel_length

def vanWinkle(Ls, mul, mus, gamma):
  recip_mu = 1/mul - (1/mul - 1/mus)*np.exp(-Ls/gamma)
  return 1/recip_mu

def objective_function(params, args):
  mul, mus, gamma = params
  data, Ls = args
  sse = np.sqrt((vanWinkle(Ls, mul, mus, gamma)-data)**2)
  return np.sum(sse)
                       

cm_per_px = gel_length/len(positions)
positions_cm = positions*cm_per_px
peak_positions_cm = peak_locations*cm_per_px
velocities = peak_positions_cm/run_time
mobilities = velocities/E_field

args = [mobilities, ladder_sizes_bp]
params = [1e-6, 4e-6, 2000]
xopt = opt.minimize(objective_function, params, method="Nelder-Mead",args=(args,))

plt.xscale("log")
for x,y in zip(ladder_sizes_bp, mobilities):
    label = x
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
plt.scatter(ladder_sizes_bp, mobilities)
plt.plot(ladder_sizes_bp,vanWinkle(ladder_sizes_bp, xopt.x[0], xopt.x[1], xopt.x[2]))
plt.xlabel('log length [bp]')
plt.ylabel('mobility [cm^2/V/s]')
plt.show()

bp_to_mobility = {"mul":xopt.x[0], "mus":xopt.x[1], "gamma":xopt.x[2]}
```

## simulate an underlying mass distribution of molecules

```python
from scipy.stats import norm

stdev = 1 #bp
list_of_lengths_bp = np.array([20, 30, 40, 50, 60, 70,80,90,100,110,120,130,140,150,160,170,180,190,200]) #bp
list_of_amounts_ng = np.array([10, 10, 10, 10, 10, 10,10,10,10, 10, 10, 10 ,10 ,10 ,10 ,10 ,10 ,10 ,10]) #ng
x = np.array(range(1,300))
size_distribution = np.zeros((len(x)))

for band in range(0, len(list_of_lengths_bp)):
  size_distribution += norm(list_of_lengths_bp[band], stdev).pdf(x)*list_of_amounts_ng[band]
# pdf += norm(500, 1).pdf(x)
# plt.xscale("log")
plt.plot(x, size_distribution)
plt.xlabel('length [bp]')
plt.ylabel('ng')
plt.show()

```

## simulate migration in gel and final convoluted distribution

```python
intensity_profile = np.zeros((len(positions)))
position_conversion = len(intensity_profile)/gel_length
for i in range(0, len(size_distribution)):
  if size_distribution[i] > 1e-5:
    band_intensity = ng_to_intensity["m"]*size_distribution[i]**0.5 + ng_to_intensity["b"]
    band_mobility = vanWinkle(i, bp_to_mobility["mul"], bp_to_mobility["mus"], bp_to_mobility["gamma"])
    band_velocity = E_field*band_mobility #seconds
    band_distance_traveled = band_velocity*200*60 #cm
    band_position = int(position_conversion*band_distance_traveled)
    intensity_profile += band_intensity*norm(band_position, (conv_model.coef_*band_position + conv_model.intercept_)[0]).pdf(positions)

    

# plt.xlim(200,300)
plt.plot(positions, intensity_profile)
plt.xlabel('position [px]')
plt.ylabel('intesity [a.u.]')
plt.show()
gel_sim = np.zeros((len(positions), 100))
for i in range(len(intensity_profile)):
  gel_sim[i,:] = intensity_profile[i]
plt.imshow(gel_sim,cmap="Greys_r")  
```

## set up transfer matrix for deconvolution

```python
x = np.array(range(1,300))
T_mat_ = np.zeros((len(x),len(positions)))
for j in range(0,len(x)):
  size_distribution = norm(j, 0.001).pdf(x)*10 #10 ng standard
  intensity_profile = np.zeros((len(positions)))
  position_conversion = len(intensity_profile)/gel_length
  for i in range(0, len(size_distribution)):
    if size_distribution[i] > 1e-5:
      band_intensity = ng_to_intensity["m"]*size_distribution[i]**0.5 + ng_to_intensity["b"]
      band_mobility = vanWinkle(i, bp_to_mobility["mul"], bp_to_mobility["mus"], bp_to_mobility["gamma"])
      band_velocity = E_field*band_mobility #seconds
      band_distance_traveled = band_velocity*run_time #cm
      band_position = int(position_conversion*band_distance_traveled)
      intensity_profile += band_intensity*norm(band_position, (conv_model.coef_*band_position + conv_model.intercept_)[0]).pdf(positions)  
  T_mat_[j] = intensity_profile/np.sum(intensity_profile)
plt.imshow(T_mat_)
plt.xlabel('length [bp]')
plt.ylabel('intesity [a.u.]')
```
